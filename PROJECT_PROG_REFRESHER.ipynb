# PERSONAL EXPENSE TRACKER - PROJECT 01 - Programming Refresher

```python
# import csv
import csv
# Function to add an expense
def add_expense(expenses):
    date = input("Enter the date (YYYY-MM-DD): ")  # Get the date from user
    category = input("Enter category (e.g., Food, Travel, Clothing etc. ): ")  # Get expense category
    amount = float(input("Enter the amount spent: "))  # Get amount spent and convert to float
    description = input("Enter a brief description: ")  # Get description of expense
    
    # Append the expense as a dictionary to the list
    expenses.append({'date': date, 
                     'category': category,
                     'amount': amount,
                     'description': description})
    print("Expense added successfully!!!")

# Function to view all expenses
def view_expenses(expenses):
    if not expenses:
        print("No expenses recorded.")  # Handle case when no expenses exist
        return
    
    # Loop through and display all expenses
    for exp in expenses:
        print(f"{exp['date']} | {exp['category']} | ${exp['amount']} | {exp['description']}")

# Function to set the monthly budget
def set_budget():
    return float(input("Enter your monthly budget: "))  # Get budget amount and convert to float

# Function to track budget status
def track_budget(expenses, budget):
    total_spent = sum(exp['amount'] for exp in expenses)  # Sum all expense amounts
    remaining = (budget - total_spent)  # Calculate remaining budget

    print(f"\nTotal Spent: Rs. {total_spent}")
    print(f"\nRemaining Budget: Rs. {remaining}")
    if remaining < 0:
        print("Warning: You have exceeded your budget !!!!")  # Alert if budget exceeded

# Function to save expenses to a CSV file
def save_expenses(expenses):
    with open("expenses.csv", "w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames = ['date', 'category', 'amount', 'description'])
        writer.writeheader()  # Write CSV headers
        writer.writerows(expenses)  # Write expense records
    print("Expenses saved to file.")

# Function to load expenses from a CSV file
def load_expenses():
    expenses = []
    try:
        with open("expenses.csv", "r") as file:
            reader = csv.DictReader(file)
            expenses = [{'date': row['date'],'category': row['category'],'amount': float(row['amount']),'description': row['description']} for row in reader]  # Read and convert data
    except FileNotFoundError:
        pass  # Ignore if file doesn't exist yet
    return expenses

# Function to display the menu and handle user input
def menu():
    expenses = load_expenses()  # Load previous expenses from file
    budget = 0.0  # Initialize budget
    
    while True:
        print("\nPERSONAL EXPENSE TRACKER")
        # Display menu options
        print(" 1. Add Expense: ")
        print(" 2. View Expenses: ")
        print(" 3. Set Budget: ")
        print(" 4. Track Budget: ")
        print(" 5. Save and Exit: ")
        
        choice = input("Choose an option from above(1-5): ")
        
        # Handle user choice
        if choice == "1":
            add_expense(expenses)
        elif choice == "2":
            view_expenses(expenses)
        elif choice == "3":
            budget = set_budget()
        elif choice == "4":
            track_budget(expenses, budget)
        elif choice == "5":
            save_expenses(expenses)
            print("Goodbye!")
            break  # Exit the loop
        else:
            print("You have chosen a wrong option,please try again.")  # Handle invalid input
            

# Run the program if executed directly
if __name__ == "__main__":
    menu()
```
